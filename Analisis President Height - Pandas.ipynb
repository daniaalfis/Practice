```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn; seaborn.set()  # set plot style
```

```python
"import dataset"
heights = pd.read_csv("C:/Users/yodi.prayodya/Downloads/president_heights.csv")
```

```python
heights
```

```python
height_data = np.array(heights["height(cm)"])
```

```python
height_data
```

```python
"Menghitung statistics menggunakan NumPy aggregate functions"
height_data = np.array(heights["height(cm)"])
print("Mean height:       ", height_data.mean())
print("Standard deviation:", height_data.std())
print("Minimum height:    ", height_data.min(), "order: ", height_data.argmin())
print("Maximum height:    ", height_data.max(), "order: ", height_data.argmax())
print("25th percentile:   ", np.percentile(height_data, 25))
print("Median:            ", np.median(height_data))
print("75th percentile:   ", np.percentile(height_data, 75))
```

```python
%matplotlib inline
plt.hist(height_data)
plt.title('Height Distribution of US Presidents')
plt.xlabel('height (cm)')
plt.ylabel('number');
```
